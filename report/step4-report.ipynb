## 1. Insights from EDA

The EDA showed that the target class is extremely imbalanced, with fraud making up only a tiny share of the data. It also showed that fraud is concentrated in a small number of transaction types, especially `TRANSFER` and `CASH_OUT`, while `DEBIT` and `PAYMENT` had no fraud cases at all. In addition, several monetary variables were strongly right-skewed, and the multivariate analysis suggested a fraud signature tied to balance changes, transaction amount, and time of day.

## 2. Dropping and keeping columns

I dropped `nameOrig` and `nameDest` because they are identifier-like fields that do not describe transaction behavior directly and could encourage the model to memorize account IDs instead of learning general fraud patterns. I also dropped `isFlaggedFraud` because it was extremely sparse and not a reliable predictor in this dataset. On the other hand, I kept the balance variables and amount because EDA showed they were strongly related to fraud, and I kept transaction type because it clearly separated fraud from non-fraud. 

## 3. Hyperparameter tuning strategy

I used random search with `RandomizedSearchCV`, not grid search. That choice made sense because it let me explore a reasonable range of hyperparameters without trying every possible combination, which is more efficient when tuning several models and especially useful when the search space is large. I also kept the tuning logic inside a helper function so I could reuse the same workflow across Logistic Regression, Random Forest, and XGBoost.

## 4. Performance after tuning

The tuned models performed only slightly better than the baselines because the starting performance was already extremely strong. Logistic Regression reached an F1 score of 0.9753 on the all-types data, Random Forest improved to 0.9988, and XGBoost reached 0.9982, which suggests the feature space is highly separable. On the fraud-only subset, Logistic Regression reached 0.9906 and XGBoost reached 0.9960, again showing that the signal was already very clear before tuning.

## 5. Final F1 score

The final F1 score from the fraud-only XGBoost model was 0.9960. Random Forest on the all-types dataset had the highest F1 score at 0.9988, but the fraud-only XGBoost result was also extremely strong. 

## Overall takeaway

EDA revealed a very strong fraud signal, especially in transaction type, amount, timing, and balance behavior. That made even the baseline models perform unusually well, so the models were more about confirming and slightly refining that pattern.
